# Pour mieux comprendre

## Qu'est-ce qu'un perceptron ?

Aux origines des réseaux de neurones, on retrouve le modèle du Perceptron, sur lequel les modèles plus complexes se sont basés par la suite.

L'idée est assez simple : nous avons d'un côté plusieurs neurones d'entrée, chacune stockant une information numérique. A chacune de ces neurones d'entrée est associé un poids, que le perceptron peut faire varier pour améliorer ses performances. Pour calculer la valeur de son neurone de sortie, le perceptron fait simplement la somme des valeurs des neurones d'entrée, pondérée par leurs poids, et applique au résultat une fonction appelée fonction d'activation.

![image](drive/MyDrive/datascience/img/perceptron.png)

## Qu'est-ce qu'un réseau de neurones profond ?

On peut voir les réseaux de neurones profonds comme une évolution du perceptron : au lieu de simplement avoir des neurones d'entrées et des neurones de sortie, un réseau de neurones profond va rajouter de nouvelles couches de neurones, dites **cachées**. Elles permettent d'appliquer des traitements bien plus complexes entre les entrées et les sorties, permettant d'avoir de bien meilleurs résultats.

Les couches cachées que nous voyons sur l'image ci-dessous sont simplement des couches `Fully Connected`: chaque neurone de la couche précédente est connectée à toutes les neurones de la couche suivante. Cependant, il existe bien d'autres types de couches que l'on peut appliquer entre les entrées et les sorties d'un réseau de neurones profond: nous verrons au fur et à mesure de notre étude celles qui nous serons utiles.

![image](drive/MyDrive/datascience/img/deep-neural-network.png)

## Enjeux

Le but de notre modèle sera de décrire une photographie, en analysant son contenu. Cependant, cela implique une première chose: il faut s'assurer que l'image entrée soit bien une photographie.

En effet, TouNum travaillant sur de nombreux types de documents, on peut trouver dans un dossier pêle-mêle des photos, des dessins, des documents textuels, des schémas ou encore des peintures... Nous devons donc, avec un modèle de deep learning, filtrer les images passées en entrée pour ne garder que celles correspondant bien à une photographie.

## Principe théorique

### Le problème des images

Comme nous l'avons vu plus tôt, il est possible dans un réseau de neurones profond d'appliquer un ensemble de couches fully connected entre les entrées et les sorties de notre modèle. Cependant, dans le cas d'une image, cela pose de nombreux problèmes.

Il faut imaginer une image comme un ensemble de pixels ; chaque pixel est représenté par une valeur numérique (3 valeurs dans le cas d'images en couleurs ; mais nous resterons sur l'exemple des images monochromes pour expliquer le principe). Chaque pixel de notre image correspondra donc à un neurone d'entrée dans notre modèle.

Cependant, bien que chaque pixel de notre image soit une entrée, chacun d'entre eux pris séparément ne nous donne aucune information sur l'image. En effet, lorsque l'on analyse une image par deep learning, que ce soit pour la classifier, la commenter, ou encore la transformer, on recherche les parties de l'image qui la rende unique : le bec d'un oiseau, la structure du visage d'une personne, la couleur d'un renard... On appelle ces parties pertinentes de l'image des `features` (fonctionnalités).

Ces fonctionnalités ne sont donc pas des pixels isolés ; mais bien un ensemble de pixels, dont la position les uns par rapport aux autres est très importante. Or, dans le cas des couches fully connected, on perd complètement cette spatialité des pixels dans l'image : chaque neurone (et de fait pixel) est connecté à tous les neurones de la couche suivante !

### Qu'est-ce que la convolution ?

La convolution consiste à appliquer une fenêtre (appelé noyau de convolution) sur chaque partie de l'image. Ce filtre va multiplier les valeurs de chaque neurone de la partie de l'image qu'elle traite par ses propres valeurs. On prend ensuite la somme de ces nouvelles valeurs, et on obtient la valeur d'un **seul et unique pixel** dans la nouvelle matrice générée. On décale ensuite la zone d'application du filtre d'un pixel, et on répète l'opération.

Dans l'exemple ci-dessous, on applique un noyau de taille 3x3 en forme de X (les diagonales valent 1, et le reste 0). On commence par l'appliquer sur le coin haut gauche, ce qui nous donne la valeur 4. On déplace ensuite le noyau, et on répète l'opération.

![image](drive/MyDrive/datascience/img/convolution.png)

### Qu'est-ce que le pooling ?

Le problème de n'appliquer que des convolutions, c'est que l'on est très sensible au zoom dans l'image que l'on analyse. En effet, si l'on recherche le visage de quelqu'un, mais que celui-ci est en arrière-plan, il sera difficile de l'analyser avec la même couche convolutive qu'une autre image dans laquelle il est au premier plan.

C'est pour cela que l'on a recours au `pooling` : cela consiste simplement à réduire les dimensions de l'image, ce qui permet ensuite d'appliquer une autre convolution sur une image plus petite.

Dans l'exemple ci-dessous, c'est un pooling par valeur maximum qui est appliqué : on divise l'image par zones de 4x4 pixels, et on ne retient pour chacune d'entre elles que le pixel dont la valeur est la plus élevée.

![image](drive/MyDrive/datascience/img/pooling.png)

### Réseaux de neurones convolutifs (CNN)

Les réseaux de neurones convolutifs sont ainsi constitués de plusieurs couches de convolution, entre lesquelles on applique des couches de pooling. Ces couches de convolution permettent de ne retenir de l'image que les fonctionnalités pertinentes.

![image](drive/MyDrive/datascience/img/cnn.png)

### Et dans le cas de la classification ?

Le CNN dont nous avons parlé plus tôt nous permet de transformer nos inputs, un ensemble de neurones, en un ensemble de fonctionnalités importantes dans l'image. Par la suite, il nous faut traiter ces nouvelles informations pour en déduire la classe à laquelle appartient notre image.

Pour cela, rien de mieux qu'un ensemble de couches fully connected, entrainées pour classifier nos images en fonctions des features présentes à l'intérieur !

![image](drive/MyDrive/datascience/img/classifier.png)

Après de multiples recherches, nous avons fini par trouver la source de cette erreur: comme on le voit dans le message d'erreur, elle intervient au moment de décoder une image. Tout simplement : notre dataset contient une, ou plusieurs images corrompues !

Nous avons donc écrit un algorithme qui va décoder chaque image une à une, et supprimer les images qui sont corrompues.

````python
import glob
from keras import Sequential, Input, Model
from keras.layers import RandomFlip, RandomRotation, RandomZoom, MaxPooling2D, Flatten, Dense, Dropout, UpSampling2D
import keras_tuner as kt
from keras.optimizers import RMSprop
from tensorflow import add
from tensorflow.python.data import AUTOTUNE
import IPython
import tensorflow as tf
from keras.layers import Conv2D
from tensorflow.keras.layers import *
from tensorflow.keras.models import *
from tensorflow.keras.losses import *
from tensorflow.keras.optimizers import *
from tensorflow.keras.metrics import *
from tensorflow.keras import regularizers
from tensorflow.keras.preprocessing.image import load_img
import pathlib
import numpy as np
from skimage.util import random_noise
import os
import matplotlib.pyplot as plt
import random

tf.get_logger().setLevel('ERROR')
````

````python
def scan_corrupted_images():
    # On parcourt chaque dataset
    for dataset_name in ['multiclass', 'binary']:
        # Pour chacun d'entre eux, on récupère les chemins absolus vers l'ensemble des fichiers présents dans les sous-dossiers (Painting, Photos, ...)
        img_paths = glob.glob(os.path.join('drive/MyDrive/datascience/dataset_1_' + dataset_name, '*/*.*'))
        # On initialise le tableau qui contiendra tous les chemins vers des fichiers corrompus
        corrupted_files = []

        print('Scanning ' + dataset_name + ' dataset...')
        # On demande à Keras de lire, puis de décoder chaque fichier.
        for image_path in img_paths:
            try:
                img_bytes = tf.io.read_file(image_path)
                decoded_img = tf.io.decode_image(img_bytes)
            # Si une erreur est levée, cela signifie que le fichier est corrompu: on l'ajoute à la liste.
            except tf.errors.InvalidArgumentError as e:
                print(f"-> Found 1 corrupted file: {image_path}")
                corrupted_files.append(image_path)

        print('-------------------------')
        print(f"Total: {len(corrupted_files)} corrupted files.")

        # Il suffit maintenant de parcourir les fichiers corrompus, et de les supprimer pour éviter toute erreur durant notre processus d'apprentissage.
        for bad_path in corrupted_files:
            os.remove(bad_path)
            print('-> Removed ' + bad_path)
        print()
````

### Première solution : classification binaire

Cette solution se basera sur un dataset re-labellisé, que nous avons nommé `dataset_1_binary`.
Dans un premier temps, définissons les paramètres généraux de notre dataset:

````python
image_height = 299
image_width = 299
batch_size = 32
````

Et importons notre dataset, en prenant soin de le diviser en deux parties :
- 80% des images serviront d'entraînement à notre modèle ;
- 20% d'entre-elles serviront à le valider.

````python
binary_data_dir = pathlib.Path('drive/MyDrive/datascience/dataset_1_binary')

# Import du dataset d'entraînement
not_prefetch_train_set_binary = tf.keras.preprocessing.image_dataset_from_directory(
    binary_data_dir,
    validation_split= 0.2,
    subset = "training",
    seed=42,
    image_size=(image_width, image_height),
    batch_size=batch_size,
    label_mode='binary',
)

# Import du dataset de validation
test_set_binary = tf.keras.preprocessing.image_dataset_from_directory(
    binary_data_dir,
    validation_split= 0.2,
    subset = "validation",
    seed=42,
    image_size=(image_width, image_height),
    batch_size=batch_size,
    label_mode='binary',
)
````

Maintenant que le dataset est important, il est très important de le préparer pour qu'il soit interprétable par notre modèle :
- Nous le normalisons, pour transformer toutes les données en valeurs comprises entre 0 et 1,
- Nous autorisons le traitement en parallèle des données du dataset, avec un nombre d'appels parallèles défini automatiquement selon la configuration de la machine qui exécutera le modèle,
- Nous chargeons les données en cache pour améliorer les performances lors du futur entraînement de notre modèle.

````python
def process(image,label):
    image = tf.cast(image/255. ,tf.float32)
    return image,label

train_set_binary = not_prefetch_train_set_binary\
    .map(process, num_parallel_calls=AUTOTUNE)\
    .prefetch(buffer_size=batch_size)

test_set_binary = test_set_binary\
    .map(process, num_parallel_calls=AUTOTUNE)\
    .prefetch(buffer_size=batch_size)
````

Désormais, nous pouvons déclarer une fonction qui sera responsable de construire notre modèle. Nous lui passons en paramètres les hyperparamètres choisis par Keras Tuner et configurons l'ensemble des valeurs possibles pour chaque hyperparamètres, avant de construire un modèle séquentiel en les utilisant.

````python
def model_builder(hp):

    # Data Augmentation hyperparameters tuning
    augmentation_flip = hp.Choice(        "augmentation_flip",     ['horizontal', 'vertical', 'horizontal_and_vertical'])
    augmentation_rotation = hp.Float(name='augmentation_rotation', min_value=0.2, max_value=0.8)
    augmentation_zoom = hp.Float(    name='augmentation_zoom',     min_value=0.3, max_value=0.8)

    # Convolution hyperparameters tuning
    convolution_filters=hp.Int( name='convolution_filters',    min_value=2, max_value=4)
    convolution_kernel=hp.Int(  name='convolution_kernel',     min_value=2, max_value=4),
    convolution_activation=hp.Choice('convolution_activation', ['elu', 'relu', 'sigmoid', 'tanh'])
    convolution_padding=hp.Choice(   'convolution_padding',    ['same', 'valid'])

    # Pooling hyperparameters tuning
    pooling_kernel=hp.Int( name='pooling_kernel',  min_value=2, max_value=4)
    pooling_padding=hp.Choice(  'pooling_padding', ['same', 'valid'])

    # Hidden layers hyperparameters tuning
    hidden_layers =          hp.Int(name="hidden_layers",            min_value=1, max_value=4)
    hidden_layers_neurons =  hp.Choice(  'hidden_layers_neurons',    [128, 256, 512])
    hidden_layers_activation=hp.Choice(  'hidden_layers_activation', ['elu', 'relu', 'sigmoid', 'tanh'])

    # Dropout hyperparameter tuning
    dropout = hp.Float(name='dropout', min_value=0.3, max_value=0.8)

    model = Sequential()

    # On effectue une data augmentation : on rajoute des images dans notre dataset en transformant nos images de base
    model.add(RandomFlip(mode=augmentation_flip))
    model.add(RandomRotation((-augmentation_rotation, augmentation_rotation)))
    model.add(RandomZoom((0.2, augmentation_zoom)))

    # On effectue nos couches de convolution.
    filter = 16
    for convolution_filter in range(convolution_filters):
        model.add(Conv2D(filter,
                         convolution_kernel[0],
                         activation=convolution_activation,
                         padding=convolution_padding))
        model.add(MaxPooling2D(pooling_kernel,
                         padding=pooling_padding))
        filter*=2

    # On transforme la sortie en une dimension
    model.add(Flatten())

    # Et nous ajoutons des couches cachées Fully Connected pour traiter le résultat de la convolution.
    for _ in range(hidden_layers):
        model.add(Dense(hidden_layers_neurons, activation=hidden_layers_activation))

    model.add(Dropout(dropout))

    return model
````

Le modèle ci-dessus ne contient volontairement pas de couche de sortie; celle-ci diffère selon l'approche choisie, donc nous laissons le soin à la fonction compilant le modèle (ci-dessous) de la rajouter (cela évitera la duplication de code).

````python
def binary_model_builder(hp):
    output_activation = hp.Choice('output_activation', ['elu', 'relu', 'sigmoid', 'tanh'])
    loss_function     = hp.Choice('loss_function',     ['binary_crossentropy', 'hinge', 'squared_hinge'])
    optimizer         = hp.Choice('optimizer',         ['adam', 'adamax', 'sgd'])

    model = model_builder(hp)
    model.add(Dense(1, activation=output_activation))
    model.compile(loss=loss_function,
                  optimizer=optimizer,
                  metrics='accuracy')
    return model
````

Par la suite, nous déclarons notre `Keras Tuner`, en lui préciisant son objectif: en 20 itérations de maximum 15 époques chacune, il doit maximiser la précision sur le dataset de validation.

````python
binary_tuner = kt.Hyperband(hypermodel = binary_model_builder,
                     objective = kt.Objective("val_accuracy", direction="max"),
                     max_epochs = 15,
                     hyperband_iterations=20,
                     project_name='binary_classification',)
````

Bien entendu, il faut réinitialiser le modèle après chaque itération; sinon, il serait impossible de comparer les itérations entre elles, puisque chacune bénéficierait de l'entraînement précédent. Nous déclarons donc un callback se chargeant de cette tâche :

````python
class ClearTrainingOutput(tf.keras.callbacks.Callback):
    def on_train_end(*args, **kwargs):
        IPython.display.clear_output(wait = True)
````

Et nous lançons la recherche des hyperparamètres les plus influents ! Nous passons tous les arguments nécessaires à invoquer la fonction `fit()` du modèle; ici, les datasets d'entraînement et de validation de l'approche binaire, le nombre d'époques, le callback déclaré précédemment, etc.

````python
binary_tuner.search(x=train_set_binary,
             validation_data=test_set_binary,
             epochs=15,
             shuffle=True,
             verbose=1,
             initial_epoch=0,
             callbacks=[ClearTrainingOutput()],
             use_multiprocessing=True,
             workers=6,
             batch_size=batch_size)
````

Une fois la recherche effectuée, nous extrayons les hyperparamètres les plus influents, et construisons un modèle les incluant :

````python
binary_best_params=binary_tuner.get_best_hyperparameters(num_trials=1)[0]
binary_model = binary_tuner.hypermodel.build(binary_best_params)
````

Nous déclarons une petite fonction responsable de lancer l'entraînement du modèle, et d'afficher les graphes comparant la précision et l'échec entre le dataset d'entraînement et celui de validation.

````python
def train_model(model, train_dataset, test_dataset, epochs):
    history = model.fit(train_dataset, validation_data=test_dataset, epochs=epochs, verbose=1)

    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']

    loss = history.history['loss']
    val_loss = history.history['val_loss']

    epochs_range = range(epochs)

    plt.figure(figsize=(16, 8))
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.legend(loc='lower right')
    plt.title('Training and Validation Accuracy')

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title('Training and Validation Loss')
    plt.show()
````

Nous obtenons les hyperparamètres suivant :
- Concernant l'augmentation de données :
    - Un flip uniquement vertical
    - Une rotation maximum de 50°
    - Un zoom maximum de 0.5
- Concernant la convolution :
    - 4 couches de convolution (16, 32, 64, 128 filtres)
    - Un noyau de convolution de 2x2 pixels
    - `elu` comme fonction d'activation
    - Application d'un padding
- Concernant le pooling :
    - Un noyau de pooling de 2x2 pixels
    - Application d'un padding
- Concernant les couches denses :
    - Une seule couche dense de 256 neurones
    - `relu` comme fonction d'activation
    - Un dropout de 60%
- Concernant la compilation du modèle :
    - `sigmoid` comme fonction d'activation de sortie
    - `binary_crossentropy` comme fonction d'échec

# Livrable 2 - Traitement des images

## Contexte

Dans une grande base de données d'images, il y toujours des images inutilisables, car les images sont de mauvaise qualité ou des parasites sont présents dessus.

L'objectif de ce livrable est donc de pouvoir **débruiter** des images données en entrée. C'est-à-dire enlever les parasites de l'image pour qu'elle puisse être lisible et utilisable par la suite.


Pour cela, nous allons créer et entraîner un modèle d'"**auto-encodeur**" pour retirer le bruit d'un dataset d'images.

Un auto-encodeur est un modèle d'apprentissage **non supervisé**, ce qui veut dire que l'on n'aura aucun label pour savoir s'il a raison ou non. Pour s'entraîner, il va **comparer** avec une version non bruitée de la même image.

Le principe de l'auto-encodeur est simple, il est composé d'un **encodeur** et d'un **décodeur**.

L'**encodeur** va réduire la taille de l'image à l'aide de nombreuses couches de convolution et de pooling afin d'avoir plus que l'essence même de l'image, appelée "**espace latent**".

Et le **décodeur** va faire l'inverse, il va récupérer l'essence de l'image que l'encodeur a produit et il va l'agrandir également à l'aide de nombreuses couches de convolutions et d'upsampling.

Cela va permettre de garder que les caractères essentiels de l'image et donc **supprimer** les pixels inutiles, c'est-à-dire le "bruit".

On obtient donc une image dite **reconstruite**.


## Initialisation des données

On importe toutes les bibliothèques et les paramètres utiles au projet dès le début du notebook.

````python
# Paramètres globaux

batch_size     =  100
image_size     =  512
INFERENCE_SIZE =  256
epochs         =  3
fit_verbosity  =  1
main_dir       = 'drive/MyDrive/datascience/dataset_1_binary/Photo/'
````

Pour commencer, on va récupérer les chemins de toutes les images afin de pouvoir les charger plus tard.

````python
# On récupère les chemins de chaque image dans main_dir
all_image_paths = [main_dir+file for file in os.listdir(main_dir) if file.endswith('.jpg')]

print('Nombre total d\'images:', len(all_image_paths))
````

Pour pouvoir entraîner le modèle d'auto-encodeur, on doit diviser nos données en **deux**. Une partie dite de "**training**" qui va servir à entraîner le modèle et une autre partie de "**test**" ou de "**validation**" pour vérifier les résultats de notre entraînement.

````python
# On divise le dataset en 2, pour le training et le test
train_image_paths = all_image_paths[:6000]
test_image_paths  = all_image_paths[6000:7000]

print('Nombre total d\'images d\'entraînement:', len(train_image_paths))
print('Nombre total d\'images de test:', len(test_image_paths))
````

Lors de l'importation, les images sont traitées afin qu'elles possèdent une résolution identique. De plus, les valeurs des pixels sont **normalisées** (entre 0 et 1) pour faciliter le traitement par le modèle. La valeur d'une couleur (RVB) d'un pixel étant comprise entre 0 et 255 (soit 256 couleurs) : nous divisons par 255.

````python
#Fonction qui charge les images à partir des paths
def load_images(paths, size=image_size):
    images = []
    for path in paths:
        image = load_img(path, target_size=(size, size, 3)) #On charge l'image aux dimensions souhaitées
        image = np.array(image)/255.0 # On normalise les images entre 0 et 1
        images.append(image)
    #On retourne un tableau d'images contenant les pixels RGB
    return np.array(images)
````

On crée la fonction qui permet d'ajouter du **bruit** aux images.

On l'utilisera au moment d'entraîner le modèle afin de pouvoir **comparer** l'image bruité et original.

````python
#Fonction pour ajouter du bruit aux images
def add_noise(images):
    noise  = random_noise(images, mode='gaussian', mean=0, var=0.05)
    #noise  = random_noise(noise, mode='s&p',      amount=0.2, salt_vs_pepper=0.5)
    return noise
````

Maintenant, on va créer un **générateur** qui va regrouper les images originales et bruités dans la taille de "**batch**" choisie. Cela va faire des groupes d'images de la taille du batch qu'on enverra ensemble dans le modèle pour qu'il s'entraîne plus rapidement.

Le code ci-dessous est un "générateur". Ce qui veut dire qu'il va générer les données lorsque l'on en a besoin et pas tout stocker directement dans une variable.
Le générateur est utile quand on a un très grand nombre de données comme dans notre cas. Ça permet d'éviter que le modèle soit surchargé et qu'il s'arrête à cause d'une erreur.

````python
#Fonction qui génère un dataset des images bruitées et non bruitées à partir des chemins des images
#Les paires d'images seront regroupées dans des batchs
def datagen(paths, size=image_size, batch_size=5):
    for x in range(0, len(paths), batch_size):
        batch_paths  = paths[x:x+batch_size]
        # print(batch_paths)
        batch_images = load_images(batch_paths, size=size)
        noise_images = add_noise(batch_images)
        yield noise_images, batch_images
````

Maintenant on affiche les premières images bruitées et non bruitées.

````python
#Fonction pour afficher les n premières images
def show_images(X, n):
    plt.figure(figsize=(30, 30))
    for i in range(n):
        ax = plt.subplot(6, 6, i+1)
        img = X[i]
        plt.imshow(img)
        plt.axis("off")
    plt.show()
````

On lance le générateur pour récupérer le premier batch d'image.

## Modèle - Autoencodeur

### Configuration du modèle

Maintenant, on va mettre en place le **modèle d'auto-encodeur**.

L'auto-encodeur est constitué d'un **encodeur** et d'un **décodeur**.

La première étape est donc de créer un encodeur puis un décodeur. À la suite de cela, on va créer un nouveau modèle qui va **réunir les deux**.
Il sera donc notre auto-encodeur.

````python
# Model d'autoencodeur
image = Input(shape=(None,None,3))

# Encodeur
l1 = Conv2D(64, (3,3), padding='same', activation='relu', activity_regularizer=regularizers.l1(10e-10))(image)
l2 = Conv2D(64, (3,3), padding='same', activation='relu', activity_regularizer=regularizers.l1(10e-10))(l1)

l3 = MaxPooling2D(padding='same')(l2)
l3 = Dropout(0.3)(l3)
l4 = Conv2D(128, (3,3), padding='same', activation='relu', activity_regularizer=regularizers.l1(10e-10))(l3)
l5 = Conv2D(128, (3,3), padding='same', activation='relu', activity_regularizer=regularizers.l1(10e-10))(l4)

l6 = MaxPooling2D(padding='same')(l5)
l7 = Conv2D(256, (3,3), padding='same', activation='relu', activity_regularizer=regularizers.l1(10e-10))(l6)

#Decodeur
l8  = UpSampling2D()(l7)
l9  = Conv2D(128, (3,3), padding='same', activation='relu', activity_regularizer=regularizers.l1(10e-10))(l8)
l10 = Conv2D(128, (3,3), padding='same', activation='relu', activity_regularizer=regularizers.l1(10e-10))(l9)

l11 = add([l5,l10])
l12 = UpSampling2D()(l11)
l13 = Conv2D(64, (3,3), padding='same', activation='relu', activity_regularizer=regularizers.l1(10e-10))(l12)
l14 = Conv2D(64, (3,3), padding='same', activation='relu', activity_regularizer=regularizers.l1(10e-10))(l13)

l15 = add([l14,l2])

decoded = Conv2D(3, (3,3), padding='same', activation='relu', activity_regularizer=regularizers.l1(10e-10))(l15)

#AutoEncodeur
model = Model(image, decoded)
````

Maintenant que notre auto-encodeur est conçus, on va donc compiler le modèle avec l'**optimiseure** et la **fonction de perte** choisie.

Dans notre cas, ce sera "RMSprop" et "binary_crossentropy"

````python
#On compile le modèle
model.compile(optimizer=RMSprop(learning_rate=0.0001), loss='binary_crossentropy')
````

### Entrainement

````python
#Taille de l'image
IMAGE_SHAPE = (299, 299)

#On load à partir du path l'image
def load_images(image_path):
  img = tf.io.read_file(image_path, name = None)
  img = tf.image.decode_jpeg(img, channels=0)
  img = tf.image.resize(img, IMAGE_SHAPE)
  img = tf.keras.applications.inception_v3.preprocess_input(img)
  return img, image_path
````

Nous créons un Dataset TensorFlow où les chemins et les pixels des images sont liées, puis nous les groupons par 64.

````python
# Training list contient tous les paths uniques des images
training_list = sorted(set(all_img_vector))

# On crée un nouveau dataset contenant tous les paths uniques des images
New_Img = tf.data.Dataset.from_tensor_slices(training_list)

# On ajoute pour chaque path, sa matrice de pixels associées
New_Img = New_Img.map(load_images, num_parallel_calls = tf.data.experimental.AUTOTUNE)

# On ajoute un batch de taille 64
New_Img = New_Img.batch(64, drop_remainder=False)
````

Nous cherchons à extraire les features (soit les objets de l'image) grace au model Inception V3.

````python
# Extraire les features
img_features = {}

# tqdm --> Barre de progression
for image, image_path in tqdm(New_Img):
  # On extrait les features de l'image (Inception V3)
  batch_features            = image_features_extract_model(image)
  batch_features_flattened  = tf.reshape(batch_features, (batch_features.shape[0], -1, batch_features.shape[3]))

  # sert a créer un dictionnaire avec pour chaque path ses features
  for batch_feat, path in zip(batch_features_flattened, image_path):
    feature_path = path.numpy().decode('utf-8')
    img_features[feature_path] = batch_feat.numpy()
````

### Captions

````python
# On lie les images avec les captions
def map(image_name, caption):    
    img_tensor = img_features[image_name.decode('utf-8')]
    return img_tensor, caption
````

Maintenant nous créons un générateur, cela permet lorsque l'on a beaucoup de données (donc beaucoup de mémoire à utiliser) d'optimiser sa gestion en appelant seulement les données nécessaires.

````python
# generateur
BUFFER_SIZE = 1000
BATCH_SIZE  = 64
def gen_dataset(img, capt):
    data = tf.data.Dataset.from_tensor_slices((img, capt))
    data = data.map(lambda ele1, ele2 : tf.numpy_function(map, [ele1, ele2], [tf.float32, tf.int32]),
                    num_parallel_calls = tf.data.experimental.AUTOTUNE)  
    data = (data.shuffle(BUFFER_SIZE, reshuffle_each_iteration= True).batch(BATCH_SIZE, drop_remainder = False)
    .prefetch(tf.data.experimental.AUTOTUNE))
    return data
````

Nous devons à présent nous occuper de la gestion des annotations grâce à Tokenizer. Pour cela nous allons récupérer les mots de chaque annotations, afin d'en déduire les mots les plus utilisés pour ensuite les indexer, afin de lier un mot à un identifiant. Nos annotations ne seront donc plus une suite de mots mais une suite d'identifiants.

````python
#On ajoute <start> et <end> à chaque caption
annotations = ['<start>' + ' ' + line + ' ' + '<end>' for line in annotations]

#Création du Tokenizer
top_word_cnt = 5000
tokenizer = Tokenizer(num_words = top_word_cnt+1, filters= '!"#$%^&*()_+.,:;-?/~`{}[]|\=@ ',
                      lower = True, char_level = False, 
                      oov_token = 'UNK')

#On indexe les mots
tokenizer.fit_on_texts(annotations)
train_seqs = tokenizer.texts_to_sequences(annotations)

#On remplit chaque vecteur à la longueur maximale (selon la plus longue caption)
train_seqs_len      = [len(seq) for seq in train_seqs]
longest_word_length = max(train_seqs_len)
cap_vector          = tf.keras.preprocessing.sequence.pad_sequences(train_seqs, padding= 'post', maxlen = longest_word_length,
                                                          dtype='int32', value=0)
````

### Initialisation des données

Nous divisions à présent en deux parties : 80% pour l'entraînement et 20% pour la validation.

````python
#On divise le dataset X / Y (train / test)
path_train, path_test, caption_train, caption_test = train_test_split(all_img_vector, cap_vector, test_size = 0.2, random_state = 42)

train_dataset = gen_dataset(path_train, caption_train)
test_dataset  = gen_dataset(path_test, caption_test)

sample_img_batch, sample_cap_batch = next(iter(train_dataset))
````

## Modèle

Notre architecture modèle sera composé de :


1.   Un modèle CNN, soit notre encodeur. Convolutional Neural Network est un type de réseau neuronal servant à extraire les features importantes de notre entrée (donc les objets importants de notre image).
2.   Un modèle RNN, soit notre attention et notre décodeur. Recurrent Neural Network est un type de réseau neuronal servant à choisir les mots de notre description de sortie. Le RNN prend en entrée notre sortie de CNN étudié par l'attention. Celle-ci sert à hierarchiser les mots en fonction du développement de la phrase.



### Paramètres

````python
# Paramètres

embedding_dim = 256 
units         = 512

# top 5,000 mots +1
vocab_size      = 5001

# Calcule le nombre d'itérations dans les processus d'apprentissage et d'évaluation
train_num_steps = len(path_train) // BATCH_SIZE # Nombre de batchs dans train
test_num_steps  = len(path_test) // BATCH_SIZE  # Nombre de batchs dans test

max_length              = 31
feature_shape           = batch_feat.shape[1]
attention_feature_shape = batch_feat.shape[0]
````

### Encodeur - CNN

Nous définissions la classe de notre encodeur que l'on utilisera par la suite lors de la phase d'entraînement.

````python
# Classe de notre encodeur

class Encoder(Model):
    def __init__(self,embed_dim):
        super(Encoder, self).__init__()
        self.dense = tf.keras.layers.Dense(embed_dim) #Couche dense Relu activation
        
    def call(self, features):
        features =  self.dense(features) # Extraire les features de l'image
        features =  tf.keras.activations.relu(features, alpha=0.01, max_value=None, threshold=0)
        return features

encoder=Encoder(embedding_dim)
````

### Attention | Décodeur - RNN

Nous définissions la classe de notre attention que l'on utilisera par la suite lors de la phase de décodage.

````python
# Classe de notre attention

class Attention_model(Model):
    def __init__(self, units):
        super(Attention_model, self).__init__()
        self.W1     = tf.keras.layers.Dense(units)
        self.W2     = tf.keras.layers.Dense(units)
        self.V      = tf.keras.layers.Dense(1) # Dernière couche d'une sortie
        self.units  = units

    def call(self, features, hidden):
        hidden_with_time_axis = hidden[:, tf.newaxis]
        score                 = tf.keras.activations.tanh(self.W1(features) + self.W2(hidden_with_time_axis))  
        attention_weights     = tf.keras.activations.softmax(self.V(score), axis=1) 
        context_vector        = attention_weights * features 
        context_vector        = tf.reduce_sum(context_vector, axis=1)  
        return context_vector, attention_weights
````

Après l'attention, nous définissions la classe du décodeur appelant l'attention et que l'on utilisera de manière répétée.

````python
# Classe de notre décodeur

class Decoder(Model):
    def __init__(self, embed_dim, units, vocab_size):
        super(Decoder, self).__init__()
        self.units     = units
        self.attention = Attention_model(self.units)
        self.embed     = tf.keras.layers.Embedding(vocab_size, embed_dim)
        self.gru       = tf.keras.layers.GRU(self.units,return_sequences=True,return_state=True,recurrent_initializer='glorot_uniform')
        self.d1        = tf.keras.layers.Dense(self.units)
        self.d2        = tf.keras.layers.Dense(vocab_size)
        

    def call(self,x,features, hidden):
        context_vector, attention_weights = self.attention(features, hidden) # Nous faisons appel à l'attention
        embed = self.embed(x)
        embed = tf.concat([tf.expand_dims(context_vector, 1), embed], axis = -1)
        output,state = self.gru(embed)
        output = self.d1(output)
        output = tf.reshape(output, (-1, output.shape[2]))
        output = self.d2(output)
        
        return output, state, attention_weights
    
    def init_state(self, batch_size): # on initialise 
        return tf.zeros((batch_size, self.units))

decoder   = Decoder(embedding_dim, units, vocab_size)
````

### Assemblage

````python
features  = encoder(sample_img_batch)
hidden    = decoder.init_state(batch_size=sample_cap_batch.shape[0])
dec_input = tf.expand_dims([tokenizer.word_index['<start>']] * sample_cap_batch.shape[0], 1)

predictions, hidden_out, attention_weights = decoder(dec_input, features, hidden)
print('Feature shape from Encoder: {}'.format(features.shape)) #(batch, 8*8, embed_dim)
print('Predcitions shape from Decoder: {}'.format(predictions.shape)) #(batch,vocab_size)
print('Attention weights shape from Decoder: {}'.format(attention_weights.shape)) #(batch, 8*8, embed_dim)
````

Nous prenons l'optimiseur Adam, et comme fonction de perte "SparseCategoricalCrossentropy".

````python
optimizer   = tf.keras.optimizers.Adam(learning_rate = 0.001)  #define the optimizer
loss_object = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = True, reduction = tf.keras.losses.Reduction.NONE) #define your loss object
````

````python
def loss_function(real, pred):
    mask  = tf.math.logical_not(tf.math.equal(real, 0))
    loss_ = loss_object(real, pred)

    mask   = tf.cast(mask, dtype=loss_.dtype)
    loss_ *= mask
    #loss is getting multiplied with mask to get an ideal shape
    
    return tf.reduce_mean(loss_)
````

### Entraînement

La phase d'entraînement appelle le décodeur et l'attention, pour comparer le caption attendu au caption prédit. La perte moyenne de cette camparaison du batch sera retourné par la fonction.

Pour l'appretissage nous allons utiliser la techniuqe de "teacher forcing". Celle-ci va permettre d'obtenir des résultats plus intéressants et plus rapidement. Le principe étant de corriegr à chaque mot ce que notre modèle va prédire. Cela permet de remettre lors de la phase d'entraînement notre modèle dans le droit chemin s'il commence à donner une mauvaise prédiction.

````python
@tf.function
def train_step(img_tensor, target):
    loss      = 0 # initialisation du loss pour chaque batch
    hidden    = decoder.init_state(batch_size=target.shape[0]) # matrice de 0
    dec_input = tf.expand_dims([tokenizer.word_index['<start>']] * target.shape[0], 1)
    
    with tf.GradientTape() as tape:
        encoder_op = encoder(img_tensor) # on donne les pixels a l'encodeur

        # Teacher forcing
        for r in range(1, target.shape[1]) :
          # Nous donnons encoder_op à decoder
          predictions, hidden, _ = decoder(dec_input, encoder_op, hidden)
          loss                   = loss + loss_function(target[:, r], predictions) 
          dec_input              = tf.expand_dims(target[:, r], 1)  

    avg_loss       = (loss/ int(target.shape[1])) # Moyenne des pertes par batchs
    trainable_vars = encoder.trainable_variables + decoder.trainable_variables
    grad           = tape.gradient (loss, trainable_vars) # Descente de gradient

    # Maintenant nous allons calculer le gradient et l'appliquer à l'optimiseur pendant la backpropagation.
    optimizer.apply_gradients(zip(grad, trainable_vars))

    return loss, avg_loss
````

### Évaluation

Même principe qu'au dessus pour la phase d'évaluation.

````python
@tf.function
def test_step(img_tensor, target):
    loss = 0
    
    hidden = decoder.init_state(batch_size = target.shape[0])

    dec_input = tf.expand_dims([tokenizer.word_index['<start>']] * target.shape[0], 1)
    with tf.GradientTape() as tape:
      encoder_op = encoder(img_tensor)

      # Teacher forcing
      for r in range(1, target.shape[1]) :
        #On donne encoder_op à decoder
        predictions, hidden, _ = decoder(dec_input, encoder_op, hidden)
        loss                   = loss + loss_function(target[:, r], predictions)
        dec_input              = tf.expand_dims(target[: , r], 1)

    avg_loss       = (loss/ int(target.shape[1])) # Moyenne des pertes par batchs
    trainable_vars = encoder.trainable_variables + decoder.trainable_variables
    grad           = tape.gradient (loss, trainable_vars) # Descente de gradient

    # Maintenant nous allons calculer le gradient et l'appliquer à l'optimiseur pendant la backpropagation.
    optimizer.apply_gradients(zip(grad, trainable_vars))                      


    return loss, avg_loss
````

````python
def test_loss_cal(test_dataset):
    total_loss = 0

    for (batch, (img_tensor, target)) in enumerate(test_dataset) :
      batch_loss, t_loss = test_step(img_tensor, target)
      total_loss         = total_loss + t_loss
      avg_test_loss      = total_loss / test_num_steps

    return avg_test_loss
````

### Sauvegarde du modèle

Grâce au checkpoint (callback) nous sauvegardons le modèle.

````python
checkpoint_path = "Flickr8K/checkpoint1"
ckpt = tf.train.Checkpoint(encoder=encoder,
                           decoder=decoder,
                           optimizer = optimizer)
ckpt_manager = tf.train.CheckpointManager(ckpt, checkpoint_path, max_to_keep=5)
start_epoch  = 0

if ckpt_manager.latest_checkpoint:
    start_epoch = int(ckpt_manager.latest_checkpoint.split('-')[-1])
````

### Mise en place de l'entraînement

Maintenant que nous avons définis les outils nécessaires mais aussi les données traitées, nous pouvons entraîner notre modèle pour qu'il nous donne la meilleure description possible.

````python
loss_plot      = [] # Afficher le taux d'erreur de la partie train
test_loss_plot = [] # Afficher le taux d'erreur de la partie test
EPOCHS         = 15 # Nombre d'epoque pour le training

best_test_loss = 100 # On initialise notre taux d'erreur

for epoch in tqdm(range(0, EPOCHS)):
    start      = time.time()  # Chronomètre pour le processus d'une époque
    total_loss = 0            # initialisation de la perte

    for (batch, (img_tensor, target)) in enumerate(train_dataset): # train_dataset -> chemin et pixel de chaque image
        batch_loss, t_loss = train_step(img_tensor, target) # moyenne des loss d'un seul batch (celui en cours)
        total_loss        += t_loss   # total de toutes les moyennes de loss de tous les batch  
        avg_train_loss     = total_loss / train_num_steps # moyenne des loss de tous les batch
        
    loss_plot.append(avg_train_loss)  # sert a faire le graphe suivant des taux d'erreurs du train
    test_loss = test_loss_cal(test_dataset) # moyenne des loss de tous les batch pour la partie test
    test_loss_plot.append(test_loss) # sert a faire le graphe suivant des taux d'erreurs du test
    
    print ('Epoch: {}, the train loss is {:.3f}, & test loss is {:.3f}'.format(epoch+1, avg_train_loss, test_loss))
    print ('Time taken for the epoch {} sec\n'.format(time.time() - start))
    
    if test_loss < best_test_loss: # si le loss du test est < au meilleur loss on le remplace
        print('Test loss has been reduced from %.3f to %.3f' % (best_test_loss, test_loss))
        best_test_loss = test_loss # on definit notre loss en tant que meilleur loss
        ckpt_manager.save() # on sauvegarde le modele avec nos parametres actuels
````

Affichage de la fonction perte au fil des époques entre la partie train et test. Cela nous permet d'étudier la progression de notre modèle au cours des époques et de vérifier s'il n'est pas en Over-fitting.

````python
from matplotlib.pyplot import figure
figure(figsize=(12, 8))
plt.plot(loss_plot, color='orange', label = 'training_loss_plot')
plt.plot(test_loss_plot, color='green', label = 'test_loss_plot')
plt.xlabel('Epochs', fontsize = 15, color = 'red')
plt.ylabel('Loss', fontsize = 15, color = 'red')
plt.title('Loss Plot', fontsize = 20, color = 'red')
plt.legend()
plt.show()
````

Comme nous pouvons l'apercevoir la différence entre la courbe de test et train est relativement faible, nous pouvons en conclure que notre modèle n'est pas en over-fitting.

### Prédictions

Maintenant que notre modèle est entraîné, nous allons essayer de lui fournir des images externe au Dataset Flickr8k pour étudier la prédiction.

````python
def evaluate(image):
    attention_plot = np.zeros((max_length, attention_feature_shape))

    hidden = decoder.init_state(batch_size=1)

    temp_input     = tf.expand_dims(load_images(image)[0], 0)
    img_tensor_val = image_features_extract_model(temp_input)
    img_tensor_val = tf.reshape(img_tensor_val, (img_tensor_val.shape[0], -1, img_tensor_val.shape[3]))

    features  = encoder (img_tensor_val)

    dec_input = tf.expand_dims([tokenizer.word_index['<start>']], 0)
    result    = []

    for i in range(max_length):
        predictions, hidden, attention_weights = decoder(dec_input, features, hidden)

        attention_plot[i] = tf.reshape(attention_weights, (-1, )).numpy()

        predicted_id = tf.argmax(predictions[0]).numpy()

        result.append (tokenizer.index_word[predicted_id])

        if tokenizer.index_word[predicted_id] == '<end>':
            return result, attention_plot,predictions

        dec_input = tf.expand_dims([predicted_id], 0)

    attention_plot = attention_plot[:len(result), :]
    return result, attention_plot, predictions
````

Nous filtrons le résultat pour supprimer les balises créées par le modèle, afin d'afficher uniquement la phrase.

````python
def filt_text(text):
    filt=['<start>','<unk>','<end>'] 
    temp= text.split()
    [temp.remove(j) for k in filt for j in temp if k==j]
    text=' '.join(temp)
    return text
````
